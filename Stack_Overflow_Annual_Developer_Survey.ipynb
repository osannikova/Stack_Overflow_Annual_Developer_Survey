> 1. How many respondents completed the survey?

```python
num_respondents = df.shape[0]
print("Count of respondents who completed the survey:", num_respondents)
```

> 2. How many respondents answered all the questions?

```python
complete_responses = df.dropna(how='any').shape[0]
print("Count of respondents who answered all questions:", complete_responses)
```

```python
missing_data = df.isnull().sum()
print("Count of missing values in each question:\n", missing_data)
```

> 3. What are the values of the measures of central tendency for the respondents' experience (WorkExp)?

```python
mean_workexp = df['WorkExp'].mean()
median_workexp = df['WorkExp'].median()
mode_workexp = df['WorkExp'].mode()

print("Average experience value:", mean_workexp)
print("Median of experience:", median_workexp)
print("Mode of Experience:", mode_workexp.values)
```

> 4. How many respondents work remotely?

```python
remote_work_counts = df['RemoteWork'].value_counts()
print("Count of respondents by type of work (remote/on-site):\n", remote_work_counts)
```

> 5. What percentage of respondents programming in Python?

```python
total_respondents = df.shape[0]
python_users = df['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()

python_percentage = (python_users / total_respondents) * 100
print(f"Percentage of respondents who program in Python: {python_percentage:.2f}%")
```

> 6. How many respondents have learned to program through online courses?

```python
online_learning_count = df['LearnCodeOnline'].notna().sum()
print("Number of respondents who learned to code through online courses:", online_learning_count)
```

> 7. Among the respondents who program in Python, grouped by country, what is the average and median amount of compensation (ConvertedCompYearly) in each country?

```python
python_programmers = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

compensation_stats = python_programmers.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).reset_index()
compensation_stats.columns = ['Country', 'Average_Compensation', 'Median_Compensation']

print(compensation_stats)
```

> 8. What levels of education do the 5 respondents with the highest compensation have?

```python
top_5_respondents = df.nlargest(5, 'ConvertedCompYearly')

top_5_education_levels = top_5_respondents[['ConvertedCompYearly', 'EdLevel']]

print("Education levels of the top 5 respondents with the highest compensation:\n", top_5_education_levels)
```

> 9. In each age group, what percentage of respondents program in Python?

```python
total_respondents_by_age = df.groupby('Age').size()

python_programmers_by_age = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)].groupby('Age').size()
python_percentage_by_age = (python_programmers_by_age / total_respondents_by_age) * 100

python_percentage_df = python_percentage_by_age.reset_index(name='Python_Programming_Percentage')

print(python_percentage_df)
```

> 10. Among respondents who are in the 75th percentile in terms of compensation and work remotely, which industries are the most common?

```python
compensation_75_percentile = df['CompTotal'].quantile(0.75)
high_compensation_remote_workers = df[(df['CompTotal'] >= compensation_75_percentile) & (df['RemoteWork'] == 'Remote')]

industry_counts = high_compensation_remote_workers['Industry'].value_counts()

print(industry_counts)
```
